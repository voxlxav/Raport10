```python
!wget https://raw.githubusercontent.com/voxlxav/Raport10/refs/heads/main/AppleStore.csv
!wget https://raw.githubusercontent.com/voxlxav/Raport10/refs/heads/main/googleplaystore.csv

from csv import reader

#1. pobranie, wczytanie danych
def read_data_google():
    with open('google_wget.csv') as google:
        data=list(reader(google))
    header=data[0]
    return header, data[1:]

def read_data_apple():
    with open('applestore_wget.csv') as apple:
        data=list(reader(apple))
    header=data[0]
    return header, data[1:]

#2. usuwanie błędów (google)
def remove_wrong_data_google(dataset):
    cleaned=[row for i, row in enumerate(dataset) if i != 10472]
    return cleaned

#3. usuwanie duplikatów (google)
def remove_duplicates_google(dataset):
    reviews_max = {}
    for row in dataset:
        name = row[0]
        reviews = float(row[3])
        if name not in reviews_max or reviews > reviews_max[name]:
            reviews_max[name] = reviews

    cleaned = []
    added = set()
    for row in dataset:
        name = row[0]
        reviews = float(row[3])
        if reviews_max[name] == reviews and name not in added:
            cleaned.append(row)
            added.add(name)
    return cleaned

#4. usuwanie innych niż po angielsku
def is_english(string):
    non_ascii = 0 # Corrected variable name 'on_ascii' to 'non_ascii'
    for c in string:
        if ord(c) > 127:
            non_ascii += 1
        if non_ascii > 3:
            return False
    return True

def filter_english_apps(dataset, name_index):
    return [row for row in dataset if is_english(row[name_index])]

#5. wybieranie darmowych
def filter_free_apps(dataset, price_index, free_signature):
    return [row for row in dataset if row[price_index] == free_signature]

#6. funkcje przygotowujące dane
def prepare_google():
    header, data = read_data_google()
    data = remove_wrong_data_google(data)
    data = remove_duplicates_google(data)
    data = filter_english_apps(data, name_index=0)
    data = filter_free_apps(data, price_index=7, free_signature='0')
    return header, data

def prepare_apple():
    header, data = read_data_apple()
    data = filter_english_apps(data, name_index=1)
    data = filter_free_apps(data, price_index=4, free_signature='0.0')
    return header, data

#7. analiza częstości
def freq_table(dataset, index):
    table = {}
    total = len(dataset)

    for row in dataset:
        value = row[index]
        table[value] = table.get(value, 0) + 1

    table_percentages = {k: (v / total) * 100 for k, v in table.items()}
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    sorted_table = sorted(table.items(), key=lambda x: x[1], reverse=True)
    for genre, percent in sorted_table:
        print(f"{genre} : {percent:.2f}%")

#8. średnie
def genre_avg_ratings(dataset, genre_index, rating_index):
    genre_data = {}
    for row in dataset:
        genre = row[genre_index]
        try:
            rating = float(row[rating_index])
        except:
            rating = row[rating_index].replace(',', '').replace('+', '')
            rating = float(rating)

        if genre in genre_data:
            genre_data[genre]['total'] += rating
            genre_data[genre]['count'] += 1
        else:
            genre_data[genre] = {'total': rating, 'count': 1}

    avg_ratings = {genre: round(data['total'] / data['count'], 2) for genre, data in genre_data.items()}
    return sorted(avg_ratings.items(), key=lambda x: x[1], reverse=True)

#9. weryfikacja
official_apple = {
    "Social Networking": 71548.34905660378,
    "Photo & Video": 28441.54375,
    "Games": 22788.6696905016,
    "Music": 57326.530303030304,
    "Reference": 74942.11111111111,
    "Health & Fitness": 23298.015384615384,
    "Weather": 52279.892857142855,
    "Utilities": 18684.456790123455,
    "Travel": 28243.8,
    "Shopping": 26919.690476190477,
    "News": 21248.023255813954,
    "Navigation": 86090.33333333333,
    "Lifestyle": 16485.764705882353,
    "Entertainment": 14029.830708661417,
    "Food & Drink": 33333.92307692308,
    "Sports": 23008.898550724636,
    "Book": 39758.5,
    "Finance": 31467.944444444445,
    "Education": 7003.983050847458,
    "Productivity": 21028.410714285714,
    "Business": 7491.117647058823,
    "Catalogs": 4004.0,
    "Medical": 612.0,
}

official_google = {
    "ART_AND_DESIGN": 1986335.0877192982,
    "AUTO_AND_VEHICLES": 647317.8170731707,
    "BEAUTY": 513151.88679245283,
    "BOOKS_AND_REFERENCE": 8767811.894736841,
    "BUSINESS": 1712290.1474201474,
    "COMICS": 817657.2727272727,
    "COMMUNICATION": 38456119.167247385,
    "DATING": 854028.8303030303,
    "EDUCATION": 1833495.145631068,
    "ENTERTAINMENT": 11640705.88235294,
    "EVENTS": 253542.22222222222,
    "FINANCE": 1387692.475609756,
    "FOOD_AND_DRINK": 1924897.7363636363,
    "HEALTH_AND_FITNESS": 4188821.9853479853,
    "HOUSE_AND_HOME": 1331540.5616438356,
    "LIBRARIES_AND_DEMO": 638503.734939759,
    "LIFESTYLE": 1437816.2687861272,
    "GAME": 15588015.603248259,
    "FAMILY": 3695641.8198090694,
    "MEDICAL": 120550.61980830671,
    "SOCIAL": 23253652.127118643,
    "SHOPPING": 7036877.311557789,
    "PHOTOGRAPHY": 17840110.40229885,
    "SPORTS": 3638640.1428571427,
    "TRAVEL_AND_LOCAL": 13984077.710144928,
    "TOOLS": 10801391.298666667,
    "PERSONALIZATION": 5201482.6122448975,
    "PRODUCTIVITY": 16787331.344927534,
    "PARENTING": 542603.6206896552,
    "WEATHER": 5074486.197183099,
    "VIDEO_PLAYERS": 24727872.452830188,
    "NEWS_AND_MAGAZINES": 9549178.467741935,
    "MAPS_AND_NAVIGATION": 4056941.7741935486,
}

def verify_step12_2():
    _, apple_data = prepare_apple()
    calculated = dict(genre_avg_ratings(apple_data, -5, 5))  # zamiana na dict
    for genre in official_apple:
        calc_val = calculated.get(genre, 0)
        off_val = official_apple[genre]
        if abs(calc_val - off_val) > 0.01:
            print(f"Mismatch in {genre}: calculated {calc_val}, official {off_val}")
            return False
    return True


def verify_step13_2():
    _, google_data = prepare_google()
    calculated = dict(genre_avg_ratings(google_data, 1, 5))  # zamiana na dict
    for genre in official_google:
        calc_val = calculated.get(genre, 0)
        off_val = official_google[genre]
        if abs(calc_val - off_val) > 0.01:
            print(f"Mismatch in {genre}: calculated {calc_val}, official {off_val}")
            return False
    return True


print("Apple verification:", verify_step12_2())
print("Google verification:", verify_step13_2())
print('              ')

#wywołania
if __name__ == "__main__":
    google_header, google_data = prepare_google()
    apple_header, apple_data = prepare_apple()

    print("Google Play Genres Distribution:")
    display_table(google_data, 1)

    print("\nApple Store Genres Distribution:")
    display_table(apple_data, -5)
```
